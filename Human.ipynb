```python
#Shirel features
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import StandardScaler
import numpy as np
import matplotlib.pyplot as plt

# Assuming data_scaled_df and target are defined appropriately

target = data['Affinity (nm) IC50']

# Initialize the Random Forest Regressor
rf_regressor = RandomForestRegressor(n_estimators=100, random_state=42)

# Initialize K-Folds cross-validator
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Store cross-validation results
mse_scores = []
mae_scores = []
r2_scores = []

# Lists to store true and predicted values for plotting
true_values = []
predicted_values = []

# Initialize StandardScaler
scaler = StandardScaler()

# Perform K-Folds cross-validation
for train_index, test_index in kf.split(shirel_df):
    # Splitting data into train and test for this fold
    X_train, X_test = shirel_df.iloc[train_index], shirel_df.iloc[test_index]
    y_train, y_test = target.iloc[train_index], target.iloc[test_index]

    # Scaling the data
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    # Fit the model on the training data
    rf_regressor.fit(X_train_scaled, y_train)

    # Predict on the test data
    y_pred = rf_regressor.predict(X_test_scaled)

    # Calculate Mean Squared Error for the current fold
    mse = mean_squared_error(y_test, y_pred)
    mse_scores.append(mse)

    # Calculate Mean Absolute Error for the current fold
    mae = mean_absolute_error(y_test, y_pred)
    mae_scores.append(mae)

    # Calculate R-squared (R2) score for the current fold
    r2 = r2_score(y_test, y_pred)
    r2_scores.append(r2)

    # Store true and predicted values for plotting
    true_values.extend(y_test)
    predicted_values.extend(y_pred)

# Calculate the average Mean Squared Error across all folds
average_mse = np.mean(mse_scores)

# Calculate the average Mean Absolute Error across all folds
average_mae = np.mean(mae_scores)

# Calculate the average R-squared (R2) score across all folds
average_r2 = np.mean(r2_scores)

print(f'Average Mean Squared Error: {average_mse:.4f}')
print(f'Average Mean Absolute Error: {average_mae:.4f}')
print(f'Average R-squared (R2) Score: {average_r2:.4f}')
print(f'MSE scores: {mse_scores}')
print(f'MAE scores: {mae_scores}')
print(f'R2 scores: {r2_scores}')

# Plotting true values vs predicted values
plt.figure(figsize=(8, 6))
plt.scatter(true_values, predicted_values, color='blue', alpha=0.7)
plt.plot(true_values, true_values, color='red', linestyle='--')
plt.title('True Values vs Predicted Values')
plt.xlabel('True Values')
plt.ylabel('Predicted Values')
plt.grid(True)
plt.show()
```

# **XGBoost**

```python
import pandas as pd
import xgboost as xgb
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import LeaveOneOut
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import RFE
import numpy as np
import time

# Load data
data = pd.read_csv('/content/drive/MyDrive/human_df.csv', header=0)
columns_to_drop = ['Ligand SMILES code']
data.drop(columns_to_drop, axis=1, inplace=True)

# Separate features and target
y = data['Affinity (nm) IC50']  # Assuming the target is 'Affinity (nm) IC50'
X = data.drop(['Affinity (nm) IC50'], axis=1)

# Initialize XGBRegressor
base_model = xgb.XGBRegressor(
    n_jobs=4,
    objective='reg:squarederror',  # Use 'reg:squarederror' for regression
    use_label_encoder=False
)

# Initialize RFE with the base model and specify the number of features to select
num_features_to_select = 10  # Adjust this based on your specific needs
rfe = RFE(estimator=base_model, n_features_to_select=num_features_to_select)

# Initialize Leave-One-Out cross-validator
loo = LeaveOneOut()

# Store cross-validation results
mse_scores = []
r2_scores = []
fitting_times = []
inference_times = []

# Perform LOO-CV
for train_index, test_index in loo.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    # Initialize the scaler
    scaler = StandardScaler()

    # Fit the scaler on the training data and transform both training and test data
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    # Fit the RFE model on the scaled training data
    rfe.fit(X_train_scaled, y_train)

    # Transform both training and test data with the fitted RFE model
    X_train_rfe = rfe.transform(X_train_scaled)
    X_test_rfe = rfe.transform(X_test_scaled)

    # Train the model on the selected features
    start = time.time()
    modelfit = base_model.fit(X_train_rfe, y_train, eval_metric='rmse')
    fittingTime = time.time() - start
    fitting_times.append(fittingTime)

    # Make predictions
    start = time.time()
    predictions = modelfit.predict(X_test_rfe)
    inferenceTime = time.time() - start
    inference_times.append(inferenceTime)

    # Calculate metrics
    mse = mean_squared_error(y_test, predictions)
    mse_scores.append(mse)
    r2 = r2_score(y_test, predictions)
    r2_scores.append(r2)

# Calculate average metrics across all folds
average_mse = np.mean(mse_scores)
average_r2 = np.mean(r2_scores)
average_fitting_time = np.mean(fitting_times)
average_inference_time = np.mean(inference_times)

print('Average Mean Squared Error (MSE) = {:.4f}'.format(average_mse))
print('Average R-squared (R2) Score = {:.4f}'.format(average_r2))
print('Average Fitting Time = {:.4f} seconds'.format(average_fitting_time))
print('Average Inference Time = {:.4f} seconds'.format(average_inference_time))
```
